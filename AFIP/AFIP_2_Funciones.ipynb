```python
def Chrome():
    
    # Crea una instancia de ChromeOptions, que te permite configurar opciones específicas del navegador Chrome.
    Opciones = webdriver.ChromeOptions()

    # Agregar un argumento al ChromeOptions para ignorar errores de certificado SSL.
    Opciones.add_argument('--ignore-certificate-errors')

    # Agregar un argumento al ChromeOptions para ignorar errores SSL.
    Opciones.add_argument('--ignore-ssl-errors')

    # Crea una instancia del controlador de Chrome (webdriver.Chrome) y asigna la instancia a la variable "Driver".
    # Al crear esta instancia, se lanzará una nueva ventana de Chrome controlada por Selenium.
    Driver = webdriver.Chrome(options=Opciones)  
    
    return Driver
```

```python
def Esperar_Descarga(Ruta_Archivo):
    Timeout = 30  # Tiempo máximo de espera en segundos
    Intervalo = 2  # Intervalo de tiempo entre comprobaciones en segundos
    Tiempo_Inicial = time.time()

    while not os.path.exists(Ruta_Archivo):
        if time.time() - Tiempo_Inicial > Timeout:
            raise TimeoutError("El tiempo de espera para la descarga ha sido superado")
        time.sleep(Intervalo)
```

```python
# Ingreso a la AFIP.

def Paso_0(Driver, Usuario, Contraseña):
    
    Login = 'https://auth.afip.gob.ar/contribuyente_/login.xhtml'
    
    # Abre la página.
    Driver.get(Login)
    
    # Espera hasta que el campo "Usuario" esté presente en la página.
    WebDriverWait(Driver, 10).until(EC.presence_of_element_located((By.ID, 'F1:username'))) 
    time.sleep(2)
    
    # Encuentra el campo de entrada para el nombre de usuario.
    Campo_Usuario = Driver.find_element(By.ID, 'F1:username')

    # Envía el nombre de usuario al campo de entrada.
    Campo_Usuario.send_keys(Usuario)

    # Presiona la tecla Enter para enviar el formulario.
    Campo_Usuario.send_keys(Keys.RETURN)
    
    # Espera hasta que el campo "Contraseña" esté presente en la página.
    WebDriverWait(Driver, 10).until(EC.presence_of_element_located((By.ID, 'F1:password')))
    time.sleep(2) 

    # Encuentra el campo de entrada para la contraseña.
    Campo_Pass = Driver.find_element(By.ID, 'F1:password')

    # Envía la contraseña al campo de entrada
    Campo_Pass.send_keys(Contraseña)

    # Presiona la tecla Enter para enviar el formulario.
    Campo_Pass.send_keys(Keys.RETURN) 
    
    # ---------------------------------------------------------------------------------------------------------------------

    # Espera hasta que el link "Comprobantes en línea" esté presente en la página.
    WebDriverWait(Driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, 'h3.roboto-font.regular.p-y-0.m-y-0.h4')))
    time.sleep(2)

    # Encuentra el elemento <h3> por su clase CSS.
    Comprobantes_en_Linea = Driver.find_element(By.CSS_SELECTOR, 'h3.roboto-font.regular.p-y-0.m-y-0.h4')
    time.sleep(2)

    # Hace clic en el link Comprobantes_en_Linea.
    Comprobantes_en_Linea.click()

    # ---------------------------------------------------------------------------------------------------------------------

    # Espera hasta que haya al menos dos ventanas abiertas o hasta que pase un tiempo máximo.
    WebDriverWait(Driver, 10).until(EC.number_of_windows_to_be(2))
    time.sleep(2)

    # Obtener todas las manijas de las ventanas abiertas.
    Ventanas = Driver.window_handles

    # Cambiar al controlador de la nueva pestaña.
    Driver.switch_to.window(Ventanas[1])

    # Expresión XPath con la variable
    XPath = f"//input[@value='{Empresa}']"

    # Encuentra el botón por su clase CSS.
    Boton_Empresa = Driver.find_element(By.XPATH, XPath)

    # Hace clic en el botón de empresa.
    Boton_Empresa.click()
    
    return Driver
```

```python
# Pestaña 1: "Generar Comprobante".

def Paso_1(Driver):
    
    # Espera hasta que el botón "Generar comprobante" esté presente en la página.
    WebDriverWait(Driver, 10).until(EC.presence_of_element_located((By.XPATH, "//span[@class='ui-button-text' and text()='Generar Comprobantes']"))) 
    time.sleep(2)

    # Encuentra el botón por su clase CSS.
    Boton_Generar_Comprobante = Driver.find_element(By.XPATH, "//span[@class='ui-button-text' and text()='Generar Comprobantes']")

    # Hace clic en el botón de Generar Comprobante.
    Boton_Generar_Comprobante.click()
    
    return Driver
```

```python
# Pestaña 2: Punto de ventas y tipo de comprobante.

def Paso_2(Driver, Punto_de_Ventas, Tipo_Comprobante_Value):
    # Espera hasta que la lista desplegable esté disponible en la página.
    WebDriverWait(Driver, 10).until(EC.presence_of_element_located((By.ID, "puntodeventa"))) 
    time.sleep(1)

    # Encuentra la lista desplegable de Punto de venta por su ID.
    Lista_Punto_de_Ventas = Driver.find_element(By.ID, "puntodeventa")

    # Crea una instancia de la clase Select.
    Seleccionar_PV = Select(Lista_Punto_de_Ventas)

    XPath_PV = f"//option[contains(text(), '{Punto_de_Ventas}')]"

    # Encuentra la opción que contiene el texto parcial.
    Opcion_PV = Lista_Punto_de_Ventas.find_element(By.XPATH, XPath_PV)

    # Hace clic en la opción en Punto de Venta.
    Opcion_PV.click()


    # Espera hasta que la opción esté disponible en la página.
    WebDriverWait(Driver, 10).until(EC.presence_of_element_located((By.XPATH, "//select[@name='universoComprobante']/option[@value='2']"))) 
    time.sleep(1)

    # Encuentra la lista desplegable de Tipo de comprobante por su ID.
    Lista_Tipo_Comprobante = Driver.find_element(By.ID, "universocomprobante")

    # Crea una instancia de la clase Select.
    Seleccionar_TC = Select(Lista_Tipo_Comprobante)

    # Encuentra la opción que contiene el texto parcial.
    Seleccionar_TC.select_by_value(str(Tipo_Comprobante_Value))


    # Espera hasta que la opción esté disponible en la página.
    WebDriverWait(Driver, 10).until(EC.presence_of_element_located((By.XPATH, "//input[@value='Continuar >']"))) 
    time.sleep(1)

    # Encuentra el botón por su clase CSS.
    Boton_Continuar = Driver.find_element(By.XPATH, "//input[@value='Continuar >']")

    # Hace clic en el botón de Continuar.
    Boton_Continuar.click()    
    
    return Driver
```

```python
# Pestaña 3: Fecha y Concepto de Venta.

def Paso_3(Driver, Fecha, Tipo_Concepto_Value):
    # Espera hasta que la opción esté disponible en la página.
    WebDriverWait(Driver, 10).until(EC.presence_of_element_located((By.ID, "fc"))) 
    time.sleep(1)

    # Encuentra el elemento de entrada por su id.
    Campo_Fecha = Driver.find_element(By.ID, "fc")

    # Borra cualquier valor existente en el campo.
    Campo_Fecha.clear()

    # Escribe el valor de la variable Fecha en el campo.
    Campo_Fecha.send_keys(Fecha)


    # Encuentra la lista desplegable de Concepto por su ID.
    Lista_Concepto = Driver.find_element(By.ID, "idconcepto")

    # Crea una instancia de la clase Select.
    Seleccionar_C = Select(Lista_Concepto)

    # Encuentra la opción que contiene el texto parcial.
    Seleccionar_C.select_by_value(str(Tipo_Concepto_Value))


    # Espera hasta que la opción esté disponible en la página.
    WebDriverWait(Driver, 10).until(EC.presence_of_element_located((By.XPATH, "//input[@value='Continuar >']"))) 
    time.sleep(1)

    # Encuentra el botón por su clase CSS.
    Boton_Continuar = Driver.find_element(By.XPATH, "//input[@value='Continuar >']")

    # Hace clic en el botón de Continuar.
    Boton_Continuar.click()  
    
    return Driver
```

```python
# Pestaña 4: Condición frente al IVA y Tipo de pago.

def Paso_4(Driver, Tipo_Condicion_IVA_Value, Tipo = 'Contado'):
    # Espera hasta que la opción esté disponible en la página.
    WebDriverWait(Driver, 10).until(EC.presence_of_element_located((By.ID, "idivareceptor"))) 
    time.sleep(2)

    # Encuentra la lista desplegable de Concepto por su ID.
    Lista_Condicion_IVA = Driver.find_element(By.ID, "idivareceptor")

    # Crea una instancia de la clase Select.
    Seleccionar_CIVA = Select(Lista_Condicion_IVA)

    # Encuentra la opción que contiene el texto parcial.
    Seleccionar_CIVA.select_by_value(str(Tipo_Condicion_IVA_Value))

    
    if Tipo == 'Contado':
        # Encuentra el botón por su clase CSS.
        Boton_Contado = Driver.find_element(By.ID, "formadepago1")

        # Hace clic en el botón de empresa.
        Boton_Contado.click()

        
    # Espera hasta que la opción esté disponible en la página.
    WebDriverWait(Driver, 10).until(EC.presence_of_element_located((By.XPATH, "//input[@value='Continuar >']"))) 
    time.sleep(1)

    # Encuentra el botón por su clase CSS.
    Boton_Continuar = Driver.find_element(By.XPATH, "//input[@value='Continuar >']")

    # Hace clic en el botón de Continuar.
    Boton_Continuar.click()  
    
    return Driver
```
