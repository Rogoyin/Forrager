```python
import pandas as pd
from datetime import datetime

from openpyxl import Workbook, load_workbook
from openpyxl.utils import get_column_letter
from openpyxl.utils.dataframe import dataframe_to_rows
from openpyxl.styles import NamedStyle, Font, PatternFill, Border, Side, Alignment
from openpyxl.worksheet.dimensions import ColumnDimension, DimensionHolder
from openpyxl.worksheet.worksheet import Worksheet
```

El objetivo de este programa es tomar el .csv descargado en Payway con los ingresos con tarjeta, al que previamente llamaremos "Payway.csv" y colocaremos en J:/Mi unidad/Forraje/AFIP, y darle el formato ideal para que quede listo para AFIP.

```python
Columnas = ["Descripción","Precio","Fecha"]

# Crear un DataFrame vacío con las columnas
df = pd.DataFrame(columns=Columnas)
```

```python
# Crear un DataFrame vacío con las columnas e ignoro la primer fila que tiene cosas que no sirven
df_Payway = pd.read_csv("J:/My Drive/Forraje/AFIP/Payway.csv", skiprows=1)

df_Payway
```

```python
#Agregamos los precios
df['Precio'] = df_Payway['MONTO_BRUTO']
df['Fecha'] = df_Payway['FECHA']
```

```python
# Dataframe con los precios.

Sistema = 'J:/My Drive/Forraje/Exportar.xls'
df_Sistema = pd.read_excel(Sistema)
```

```python
# def Asignar_Descripcion(Precio):
#     if Precio <= 1000:
#         return 'Zimpi'
#     elif 1000 < Precio <= 2000:
#         return 'Dog Selection Cachorro'
#     elif 2000 < Precio <= 3000:
#         return 'Old Prince Cordero'
#     elif 3000 < Precio <= 4000:
#         return 'Agility Gato Urinary'
#     elif 4000 < Precio <= 5000:
#         return 'Royal Canin Mini Puppy'
#     elif 6000 < Precio <= 10000:
#         return 'Royal Canin Mini'
#     elif 10000 < Precio <= 15000:
#         return 'Royal Canin Gato Renal x 2 kg.'
#     elif 15000 < Precio <= 20000:
#         return 'Sabrositos Mix x 15 kg.'
#     elif 20000 < Precio <= 30000:
#         return 'Maintenance x 22 kg.'
#     elif 30000 < Precio <= 40000:
#         return 'Sabrositos Cachorro x 18 kg.'
#     elif 40000 < Precio <= 50000:
#         return 'Sabrositos Gato Pescado x 20 kg.'
#     elif 50000 < Precio <= 60000:
#         return 'Royal Canin Premium x 20 kg.'
#     elif 60000 < Precio <= 70000:
#         return 'Royal Canin Renal x 10 kg.'
#     elif 70000 < Precio <= 80000:
#         return 'Royal Canin Gatitos x 7,5 kg.'
#     else:
#         return 'Pro Plan Urinary x 15 kg.'

def Asignar_Descripcion(df, Precio):
    closest_index = (df['Precio'] - Precio).abs().idxmin()
    return df.loc[closest_index, 'Descripción']

# Aplicas la función a la columna 'Precios' y asignas el resultado a la columna 'Descripción'
df['Descripción'] = df.apply(lambda row: Asignar_Descripcion(df_Sistema, row['Precio']), axis=1)
```

```python
# Nuevo orden de columnas.
Orden = ['Fecha', 'Descripción', 'Precio']
df = df[Orden]
```

```python
# Convertir la columna 'Fecha' al formato datetime
df['Fecha'] = pd.to_datetime(df['Fecha'], format='%d/%m/%Y')
```

```python
# Ordenar según la columna 'Fecha' en orden ascendente.

df = df.sort_values(by='Fecha')
```

```python
# Volver al formato dd/mm/yyyy.

# Convertir la columna 'Fecha' de nuevo al formato 'dd/mm/yy'
df['Fecha'] = df['Fecha'].dt.strftime('%d/%m/%Y')
```

```python
# Resetear index del df.
df = df.reset_index(drop=True)
```

```python
# Configurar Pandas para mostrar todas las filas y/o columnas.
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

df
```

```python
# Guarda el DataFrame en un archivo Excel

# Define el nombre del archivo con la fecha actual utilizando una f-string
Nombre_Archivo = f"J:/My Drive/Forraje/AFIP/AFIP.xlsx"

# Guarda el DataFrame en el archivo CSV con el nombre actualizado
df.to_excel(Nombre_Archivo, index=False)
```

```python
# Cargamos Excel.
Libro = load_workbook(Nombre_Archivo)

# Seleccionar hoja activa.
Hoja = Libro.active
```

```python
# Anchos.
Hoja.column_dimensions["A"].width = 15
Hoja.column_dimensions["B"].width = 40
Hoja.column_dimensions["C"].width = 15 
```

```python
# Formato general con cero decimales para columna de precios.
General = NamedStyle(name="general")

# Decimales.
General.number_format = '0'

# Columnas.
for Columna in ["B"]:
    for Celda in Hoja[Columna]:
        Celda.style = General
```

```python
# Centrar.
Centrar = Alignment(horizontal='center', vertical='center')

# Iteración sobre cada celda de la fila.

for i in Hoja.columns:
    for Celda in i:
        Celda.alignment = Centrar
```

```python
# Guardar el archivo
Libro.save(Nombre_Archivo)
```
