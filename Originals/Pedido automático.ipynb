## Instrucciones

1. Vas a Inicio e iniciás Anaconda Navigator.
   
2. Iniciás Jupyter Notebook.
   
3. Entrás al programa 'Pedidos automáticos'.
   
4. Vas al menú de arriba, apretás Run y apretás "Run All Cells".
   
5. Abrir el archivo de Excel que se llama "Pedidos.xlsx"
   
6. Escribir todos los productos bien en la columna Descripcion, así como aparecen en el sistema.
    
7. Si no te acordás cómo se llaman, o te da pajota buscar, los podés buscar de esta forma, yendo a 'Pedidos automáticos' (o sea, acá) y escribiendo en algún cuadradito de este programa, por ejemplo, si queres buscar Pachá Mix:

    Buscar('Pach')

    Y apretas Shift + Enter. 

8. Te van a salir de resultado todos los productos que tengan 'Pach' en su nombre. Copiás y pegás en el Excel el que te interese en la columna Descripcion.
   
9. En la columna cantidad ponés la cantidad que querés comprar de cada producto.
   
10. En la columna "Proveedor de preferencia" podés agregar un proveedor al que le quieras comprar ese producto aunque no sea el más barato. Si no tenés preferencia y querés el más barato, dejala en blanco.
   
11. Guardás el archivo Excel.
    
12. Volvés acá. Si querés ver el pedido a un proveedor en particular, por ejemplo, de Bena, ponés:

    Pedido(Benavidez)

    Y apretás Shift + Enter.

13. Si querés ver los pedidos de todos los proveedores, escribís:

    Todos()


Una pelotudez. Listo el pollo, pelada la nagalli, tenés todos los pedidos listos para pasar.

14. Si querés capturar parte de la pantalla para pasarlos por Whatsapp, podés apretar Windows + Shift + S.
