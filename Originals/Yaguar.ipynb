```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import TimeoutException
import time
import pandas as pd
```

```python
# Dataframe de Yaguar.
Proveedores = 'J:/My Drive/Forraje/Proveedores.xlsx'
df = pd.read_excel(Proveedores, sheet_name="12. Yaguar")
```

```python
# Reemplazar valores no finitos con cero
df['Codigos'] = df['Codigos'].fillna(9999999999999)

# Convertir la columna a enteros
df['Codigos'] = df['Codigos'].astype(int)

# Pasar Precios a precios viejos
df['Precios viejos'] = df['Precios']

df
```

```python
# Códigos de productos.
Codigos = df['Codigos'].tolist()
```

```python
Yaguar = 'https://shop.yaguar.com.ar/frontendSP/asp/home.asp#'

Usuario = 'carolina8101924@gmail.com'
Pass = '27202147025'

Precios = []
```

```python
# Inicializar el navegador (en este caso, Google Chrome).
Driver = webdriver.Chrome()

# Abrir la página web.
Driver.get(Yaguar)

# Campo de Login.
Campo_Login = WebDriverWait(Driver, 30).until(EC.presence_of_element_located((By.NAME, "Login")))
Campo_Login.send_keys(Usuario)

# Campo de Contraseña.
Campo_Contraseña = WebDriverWait(Driver, 30).until(EC.presence_of_element_located((By.NAME, "Password")))
Campo_Contraseña.send_keys(Pass)

# Botón Enviar.
Enviar = WebDriverWait(Driver, 30).until(EC.presence_of_element_located((By.CLASS_NAME, "Enviar")))
Enviar.click()

for i in range (0, len(Codigos)):
    # Campo de búsqueda.
    Campo_Busqueda = WebDriverWait(Driver, 30).until(EC.presence_of_element_located((By.NAME, "txtBusqueda")))
    
    # Borrar cualquier texto existente en el campo de búsqueda.
    Campo_Busqueda.clear()
    
    # Escribir en el campo de búsqueda.
    Texto_Buscado = Codigos[i]
    Campo_Busqueda.send_keys(Texto_Buscado)
    
    # Presionar la tecla "Enter"
    Campo_Busqueda.send_keys(Keys.RETURN)
    
    # Cambiar al contexto del iframe.
    Iframe = Driver.find_element(By.ID, "ifrContenido")
    Driver.switch_to.frame(Iframe)

    # Agregar una pausa.
    time.sleep(1)
    
    try:
        # Precio del producto.
        Precio = WebDriverWait(Driver, 3).until(EC.presence_of_element_located((By.XPATH, "//tbody//tr//td//div//span[@class='PRODPRECIOA']")))
    
        # Obtener el contenido del precio.
        Texto_Precio = Precio.text
    
        # Agregar precio a la lista Precios.
        Precios.append(Texto_Precio)

        # Agregar una pausa.
        time.sleep(1)
    
    except TimeoutException:
        # Agregar un 0 a la lista Precios.
        Precios.append(0)

        # Agregar una pausa.
        time.sleep(1)
    
    finally:
        # Salir del contexto del iframe para volver al contexto predeterminado de la página principal
        Driver.switch_to.default_content()

        # Agregar una pausa.
        time.sleep(1)

        # Printear.
        print(Codigos[i])
```

```python
# Pasar precios a df.
df['Precios'] = Precios
```

```python
# Reemplazar valores.
df['Precios'] = df['Precios'].str.replace(".", "")
df['Precios'] = df['Precios'].str.replace(",", ".")
df['Precios'] = df['Precios'].fillna(0)
df['Precios'] = df['Precios'].astype(float)
df['Precios viejos'] = df['Precios viejos'].astype(float)
```

```python
# Reemplazar los valores de Precios con los valores correspondientes de Precios viejos si el valor en Precios es 0
df.loc[df['Precios'] == 0, 'Precios'] = df.loc[df['Precios'] == 0, 'Precios viejos']
```

```python
df['Comparación'] = df['Precios'] - df['Precios viejos']

df
```

```python
# Guarda el DataFrame en el archivo CSV con el nombre actualizado.
Nombre_Archivo = f"J:/My Drive/Forraje/Yaguar.xlsx"

df.to_excel(Nombre_Archivo, index=False)
```
