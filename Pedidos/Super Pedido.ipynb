```python
Stock = {
    "Piedritas Sanitarias Chicas x 10 u.": 4,
    "Piedritas Sanitarias Grandes x 5 u.": 1,
    "Piedritas Sanitarias x 10 kg": 2,
    "Rosco Carne x 22 kg.": 4,
    "Rosco Cocktail x 22 kg.": 2,
    "Rosco Pollo x 15 kg.": 3,
    "Rosco Carne x 15 kg.": 4,
    "Rosco Cocktail x 15 kg.": 2,
    "Arroz Gran Campeon x 15 kg.": 2,
    "Chacal x 15 kg.": 5,
    "Dr. Perrot x 20 kg.": 10,
    "Dr. Perrot x 15 kg.": 2,
    "Pacha Cocktail x 22 kg.": 3,
    "Pacha Mix x 22 kg.": 4,
    "Vagoneta Gato x 10 kg.": 1,
    "Vagoneta Gato x 20 kg.": 1
}
```

```python
Pedido = pd.DataFrame()

for Producto, Cantidad in Stock.items():
    Productos_Totales = list(df_Baratos['Descripcion'])

    if Producto in Productos_Totales:
        Producto_Final = {}
        Producto_Final['Producto'] = Producto
        Producto_Final['Stock'] = Cantidad
        Producto_Final['Proveedor'] = df_Baratos[df_Baratos['Descripcion'] == Producto]['Proveedor 1'].iloc[0]
        Producto_Final['Precio'] = df_Baratos[df_Baratos['Descripcion'] == Producto]['Precio 1'].iloc[0]
        Pedido = pd.concat([Pedido, pd.DataFrame([Producto_Final])], ignore_index=True)
    else:
        print(f"{Producto} no está en la lista.")

Pedido.sort_values(by='Proveedor', ascending=True, inplace=True)

Pedido.head(100)
```
